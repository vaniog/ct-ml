```python
import pandas as pd

data = pd.read_csv("../diabetes_dataset.csv")

print(len(data))

target_column = "Outcome"
feature_columns = data.columns.values
feature_columns = feature_columns[(feature_columns != target_column) & (feature_columns != "Unnamed: 0")]

data.head()
```

Предобработайте его: заполните пропуски, нормализуйте и т.д.

```python
from sklearn.preprocessing import StandardScaler

X = data[feature_columns].values
y = data[target_column].values

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

Разбейте на тренировочную и тестовую часть.

```python
from sklearn.model_selection import train_test_split

y = y.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(
    X_normalized, y, test_size=0.2, random_state=42
)
```

```python
from simple_neural_layer import SimpleNeuralLayer
from rbf_neural_layer import RBFNeuralLayer
from neural_network import NeuralNetwork
import numpy as np
from sklearn.metrics import f1_score

model = NeuralNetwork([SimpleNeuralLayer(16, 2), RBFNeuralLayer(2, 2), SimpleNeuralLayer(2, 1)])

model.fit(X_train, y_train, epochs=100, learning_rate=0.1, batch_size=100)
y_pred = model.predict(X_test)
y_pred = np.round(y_pred)
print(f1_score(y_test, y_pred, average='micro'))
```
